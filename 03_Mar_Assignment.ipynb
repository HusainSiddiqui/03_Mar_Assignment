Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. Scatter plot: A scatter plot is used to visualize the relationship between two continuous variables. It can help to identify any patterns, clusters, or outliers in the data.


2. Line plot: A line plot is used to visualize the trend of a continuous variable over time or some other continuous variable. It is useful for detecting any changes or patterns in the data over time.


3. Bar plot: A bar plot is used to visualize the distribution of a categorical variable or the relationship between a categorical variable and a continuous variable. It can help to identify any differences or similarities between different categories.


3. Histogram: A histogram is used to visualize the distribution of a continuous variable. It can help to identify the center, spread, and shape of the data.


4. Heatmap: A heatmap is used to visualize the relationship between two categorical variables or two continuous variables. It is useful for identifying any patterns or trends in the data.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")
print('column: ',fmri.columns)
# Create the line plot
sns.lineplot(data=fmri,x="timepoint", y="signal", hue="event", style="region")
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create the first box plot (age)
sns.boxplot(x="pclass", y="age", data=titanic)

# Create the second box plot (fare)
sns.boxplot(x="pclass", y="fare", data=titanic)
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

# Load the titanic dataset
diamond = sns.load_dataset("diamonds")

# Create the first box plot (age)
sns.histplot(data=diamond,x="price", hue="cut")
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns

# Load the titanic dataset
iris = sns.load_dataset("iris")

# Create the first box plot (age)
sns.pairplot(data=iris, hue="species")
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

# Load the flights dataset
flights = sns.load_dataset("flights")

# Reshape the data into a pivot table
flights_pivot = flights.pivot("month", "year", "passengers")

flights_pivot
```

```python
# Create the heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu")
```
